4 - Dos países que foram avaliados, qual os 5 países que tem a maior taxa de usabilidade de internet em relação a população?

```python
df3["percent_usage"] = df3["internet_users"]/df3["population"]*100
df3 = df3.sort_values(by="percent_usage", ascending=False)
df3.head(15)
```

5 - Faça um gráfico que mostre o número de países que apresentem uma taxa de usabilidade maior que 75% agrupados por Continente?


```python
df3_continental = df3[['country_region','country_name','population','internet_users']]
df3_continental = df3_continental.groupby(by=["country_region", "country_name"]).sum()

df3_continental['percent_usage'] = df3_continental["internet_users"]/df3_continental["population"]*100
df3_continental = df3_continental.sort_values(by="percent_usage", ascending=False)
df3_continental.head(50)
```
