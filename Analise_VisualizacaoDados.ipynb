```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Importante acrescentar o Pandas para a leitura e agrupamento dos dados
```

```python
# Importando o DataFrame

credit = pd.read_excel("/content/DataFrame_CreditCards.xlsx")
credit.info()
```

```python
# Agrupando os dados - Nível de Escolaridade

credit_group = credit.groupby(["nivel_escolaridade"]).mean().reset_index()

credit_group
```

# Gráficos de Barras

```python
# Gráfico MATPLOTLIB

# Média do valor de crédito concedido segundo o nível de escolaridade

plt.figure(figsize=(8,8))

plt.title("Valor de Crédito concedido segundo o Nível de Escolaridade")

plt.xlabel("Nível de Escolaridade")

plt.ylabel("Média do valor de crédito concedido")

plt.bar(credit_group["nivel_escolaridade"],
        credit_group["valor_credito"],
        color="Red");
```

```python
help(plt.bar) # Comando de Ajuda que demonstra as funcionalidades disponíveis em um determinado modelo
```

```python
# Gráfico SEABORN

# Extrato bancário dos clientes segundo o nível de escolaridade dos clientes para os 4 ultimos meses

fig, axes = plt.subplots(2,2, figsize=(15,10))

sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_jun_2005", ax = axes[0,0]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_jul_2005", ax = axes[0,1]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_ago_2005", ax = axes[1,0]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_set_2005", ax = axes[1,1]);
```

```python
help(sns.barplot)
```

```python
# Gráfico PLOTLY

# Valor pago dos clientes segundo o nível de escolaridade dos clientes para os 4 ultimos meses

fig = go.Figure()

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_abr_2005"],
                     marker_color='Red',
                     name="Valor Pago ABR/2005"))

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_mai_2005"],
                     marker_color='Blue',
                     name="Valor Pago MAI/2005"))

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_jun_2005"],
                     marker_color='Yellow',
                     name="Valor Pago JUN/2005"))

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_jul_2005"],
                     marker_color='Green',
                     name="Valor Pago JUL/2005"))

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_ago_2005"],
                     marker_color='Purple',
                     name="Valor Pago AGO/2005"))

fig.add_trace(go.Bar(x = credit_group["nivel_escolaridade"],
                     y = credit_group["valor_pag_set_2005"],
                     marker_color='Orange',
                     name="Valor Pago SET/2005"))

fig.update_layout(yaxis=dict(title="Valor do Pagamento em 2005",
                             titlefont_size=14,
                             tickfont_size=10,),
                  xaxis=dict(title="Nível de Escolaridade",
                             titlefont_size=14,
                             tickfont_size=10,))

# Salvando gráfico interativo

fig.write_html("valor_pago_escolaridade.html")

# Comando permite salvar o gráfico em HTML para exibição futura
```

```python
fig.show()
```

```python
# Gráfico PLOTLY - Subplots em Vários Gráficos


# Valor pago ao longo dos meses de acordo com a Idade

fig = go.Figure()

from plotly.subplots import make_subplots

fig = make_subplots(rows=3,cols=2, shared_yaxes=True,
                    subplot_titles=("Abril","Maio","Junho","Julho","Agosto","Setembro"))

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_abr_2005"],
                          line=dict(color='red', width=1.5,),
                          name="Valor Pago 04/2005"), rows=1, cols=1)

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_mai_2005"],
                          line=dict(color='blue', width=1.5,),
                          name="Valor Pago 05/2005"), rows=1, cols=2)

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_jun_2005"],
                          line=dict(color='green', width=1.5,),
                          name="Valor Pago 06/2005"), rows=2, cols=1)

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_jul_2005"],
                          line=dict(color='yellow', width=1.5,),
                          name="Valor Pago 07/2005"), rows=2, cols=2)

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_ago_2005"],
                          line=dict(color='orange', width=1.5,),
                          name="Valor Pago 08/2005"), rows=3, cols=1)

fig.add_traces(go.Scatter(x=credit_age["Idade"],
                          y=credit_age["valor_pag_set_2005"],
                          line=dict(color='purple', width=1.5,),
                          name="Valor Pago 09/2005"), rows=3, cols=2)

# Para estabelecer 2 linhas em um mesmo gráfico, basta utilizarmos a mesma referencia dentro do subplot

# Acrescentando mais informações

fig.update_layout(title_text="Valor pago em 2005 ao longo dos meses de acordo com a Idade dos clientes")

fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=1, col=1, titlefont_size=8)
fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=1, col=2, titlefont_size=8)
fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=2, col=1, titlefont_size=8)
fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=2, col=2, titlefont_size=8)
fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=3, col=1, titlefont_size=8)
fig.update_yaxes(title_text="Valor Pago (Dólar NT)", row=3, col=2, titlefont_size=8)

fig.update_xaxes(title_text="Idade", row=3, col=1, titlefont_size=12)
fig.update_xaxes(title_text="Idade", row=3, col=2, titlefont_size=12)

fig.show()

# Salvando o arquivo em HTML

fig.write_html("valor_pago_idade.html")
```

## Plotagem com base em algumas Funções Trigonometricas

```python
# Impotando todas as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
# Gráfico MATPLOTLIB - Plotagem Linear

# Necessário criar uma matriz - Função linspace da Numpy

t = np.linspace(0.0, 2.0, 100)
t

# Criando matrizes com diferentes funções trigonometricas

f1 = np.sin(np.pi * t) # Seno de Pi * o array t
f2 = np.sin(2 * np.pi * t) # Seno de 2 * Pi * o array t

plt.figure(figsize=(10,8))

plt.plot(t,f1, label= "Seno de Pi * t", linewidth=2.5)
plt.plot(t,f2, label= "2 Seno de Pi * t", linewidth=2.5)

plt.legend()

plt.grid(color='black', linestyle='--', linewidth=.25)

plt.title('Senoides')

; # Limpa as informações do gráfico
```

## Diagrama de Correlação

```python
# Gráfico SEABORN - Diagrama de Correlação

credit_alt = credit.drop(["Unnamed: 0","ID"],axis=1) # Removendo as colunas que não influenciam na análise

plt.figure(figsize=(20,20))

sns.heatmap(credit_alt.corr(), cmap='coolwarm', annot=True)

# O ideal é comparar variáveis de categoricas com dados quantitativos

```

**Correlação Positiva** - As variáveis podem ser diretamente proporcionais (Quando aumenta uma variável, outras podem aumentar também)

**Correlação Positiva** - As variáveis podem ser inversamente proporcionais (Quando aumenta uma variável, outras podem reduzir, efeito contrário)

```python
# Gráfico DISPERSÃO

x = [1.2, 2.5, 3.3, 4.1, 4.4, 5]
y = [65, 35, 22, 50, 18, 15]
cores = ['red', 'green', 'yellow', 'yellow', 'green', 'red']
tamanho = np.array([100, 60, 40, 80, 30, 70]) * 2

plt.scatter(
    x, y,
    s=tamanho,
    c=cores,
    alpha=.5,
    # marker='>'
    # cmap='Set1',
    # norm=
    # vmin=0,
    # vmax=1
    # linewidths=40
)

plt.xticks(np.arange(1.0,5.1,0.5))
plt.yticks(range(10,61,10))
plt.show()
```

```python
# Gráfico de Barras Verticais

x = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
y = [10, 8, 6, 4, 2, 1]

plt.bar(x, y, alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.9)
plt.ylim(0,12)
plt.ylabel('Usage')
plt.title('Programing Language Usage')
plt.show()

```

```python
# Gráfico de Barras Horizontais

x = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
y = [10, 8, 6, 4, 2, 1]

plt.barh(x, y, alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.6)
plt.xlim(0,10)
plt.xlabel('Usage')
plt.title('Programing Language Usage')
plt.show()
```

```python
# Gráfico de Barras Verticais e Horizontais em Subplots - MATPLOTLIB

fig, axes = plt.subplots(1, 2, figsize=(15,5))

axes[0].barh(x, y, alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.6)
axes[0].set_xlim(0,12)
axes[0].set_xlabel('Usage')
axes[0].set_title('Programing Language Usage')

axes[1].bar(x, y, alpha=.8, color='mediumslateblue', edgecolor='black', linewidth=.9)
axes[1].set_ylim(0,12)
axes[1].set_ylabel('Usage')
axes[1].set_title('Programing Language Usage')

plt.show()
```
